```python
#Rearrange the Array
def makeArray(Array):
    New=np.array(Array[0])

    for i in range(1,len(Array)):
        New = np.append(New,Array[i],axis=0)
        
    return New
```

```python
def readData(Stations):
    
    Data, C = [], []
    
#     print(Stations)
    file = Stations
    newfile = 'clean_'+file
    df = pd.read_csv('LMS Data2/'+file, header=0, index_col=None)
    print(file)
    df = df.reset_index(drop=True)
    data=df[['Date','Tem','Hum','Pressure','Rain','Light','Ax','Ay','Az','Wx','Wy','Wz','Moisture','Count']].values
#     data=data.astype('float32')
    count=df['Count'].values
    count=count.astype('float32')

    #Normalize the data
#     scaler = MinMaxScaler(feature_range=(0, 1))
#     data = scaler.fit_transform(data)

    Data.append(data)
    C.append(count)
           
    
        
    return makeArray(Data), makeArray(C)
```

```python
def cal_avg(Data, j):
    newData=[]
    for D in Data:
        for DD in D:
            if DD[j]!=0:
                newData.append([DD[0],DD[j]])

    new_data=np.array(newData)
    Temp={}
    for i in range(24):
        M = new_data[np.where(new_data[:,0]==i)][:,1].mean()
        
        if j!=3:
#             print(i,round(M,0))
            Temp[i]=round(M,0)
        else:
#             print(i,round(M,2))
            Temp[i]=round(M,2)

    return Temp
```
